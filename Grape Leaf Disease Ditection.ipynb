### Part A – Application area review

Using artificial intelligence in agriculture has become a major contribution to increasing productivity and effectiveness for agriculture domain specially in the tasks like crop monitoring and disease management. Grape agriculture is facing very sensitive to diseases in leaves such as leaf spot, black measles and black rot that critically affect yield and fruit quality if not detected at an early stage of grape cultivation. Traditional disease identifications take a lot of time due to identifications heavily relying on expert visual examination. Also, it is impractical for large scale vineyards. Therefore, automated grape disease detection using artificial intelligence techniques has become one of top priorities in the AI research field.

Deep learning technologies have allowed the development of intelligent systems that capable of analyzing grape leaf images and identifying disease patterns with high accuracy and low loss. Convolutional neutral networks (CNN) have shown better performance than traditional machine learning approaches by learning discriminative features automatically from raw image data.

Primitive method for grape disease studies was classical image processing techniques combined with machine learning classifiers. Handcrafted features and shape-based features were extracted from leaf images to use with classifiers like support vector machines and k-Nearest neighbors. Although these methods reached reasonable accuracy with specific environments, they had difficulties when the pictures has background complexity, variations in lighting and disease severity. (Pujrari et al., 2016).

Then introduction of CNNs noted an outstanding turning point for agriculture domain. CNN has pooling, convolutional and fully connected layers that enable hierarchical feature learning that makes them very effective for visual pattern recognition. Furthermore transfer learning CNN used to develop leveraging pretrained architectures like AlexNet, VGGNet and MobileNetV2 to enhance CNN performance. By fine tuning these models on leaf datasets with diseases information researchers have archived improved accuracy and reduced training time and computational cost. Also transfer learning CNNS like MobileNetV2 and more pretrained models has been adopted to overcome the difficulity of limited agriculture datasets and using fine tune method on pre trained models on grape leaf images researches achieved faster convergence and improved classification accuracy. (Ferentinos, 2018)

Recent studies have found methods to detect objects to localise diseased regions within grape leaf images. Faster R-CNN is a region based CNN that can use to identify and segment disease affected areas qualifying more detailed analysis of disease severeness. But that method is computationally intensive and less suitable for real time environments. (Ren et al., 2015) Single stage detectors like YOLO (You Only Look Once) have gained trust and popularity to overcome these limitations. Those models perform classification and localisation at the same time to make them suitable for real time disease detection by using smart farming systems, mobile devices and drones. (Redmon and Farhadi, 2018)

Even though agriculture domain has significant progress overtime there are still some challenges remain in automated disease detection. Limitations like similarities between disease symptoms, class imbalance and occlusion continue to affect model accuracy and performance. Also current researches may not fully reflect for real world vineyard conditions because many existing studies relying on curated datasets and laboratory-based datasets. (Too et al., 2019)

According to above information we can come to a conclusion that grrape disease detection systems based on Artificial intelligence has evolved significantly from traditional image proccessing methods to advanced deep learning models. CNNs combined with transfer learning have shown to be very effective to classify grape leaf diseases. Even though grape disease deetection has evlved to be very accurate with high performance there still some demandings to be deeveloped to get a extordinary results in this field.

**Convolutional Neural Networks (Pre Trained Model - MobileNetV2)**

CNN is inspired by how neurons process visual informaion in the human brain. CNNs currently using for image classification, computer vision and object recognition tasks because it is one of most popuar and most effective technique that can use for those tasks with high accuracy. Furthermore in genaral artificial neural networks (ANNs) perform well for many problems but CNNs can be considered as more advanced and specializes extension of ANNs specifically designed for image based data. The architecture of CNNSs typically includes an input layer, multiple convolutional layers, pooling layers, a flatten layer and then finally fully connected layers leading to the output layer. Convolutional layers used for extracting important features like edges, textures and colour patterns from images. Pooling layers used to reduce the spatial dimentions and computational complexity. Flatten layer converts the extracted feature maps to one dimentional array that proccessed by fully connected layers. After those proccesses output layer produce the final classification results. CNNs are very flexible and can be adapted to image related tasks like classification, object detection and segmentation according to the selected architecture. CNNsare speccially suitable for the graoe leaf disease detection because disease symptoms appear as visible patterns on outside of leaf as discoloration, texture changes and spots.

Pre Trained CNN model is sued in this pproject through transfer learning technique. Transfer larning works as taking a model that already trained on a general dataset like MobileNetV2 that consists many everyday photos and using what it learned for a new main problem. Pre trained models have already spent a lot of time learning new features from all those images. Therefore instead of starting from beginging and training brand new CNN to detect grape leaf diseases we take MobileNetV2 pre trained dataset to train model faster and accurately.

One of the mainn advantages of pre trained model is that it perfroms perfectly well even with smaller datasets becasue it leverages from previously learned features. It also reduces computational cost compared to trainign a CNN from start. Furthermore as results pre trained models generally get higher accuracy and better generalization for image classfication tasks. However to get an accurate results pre trained CNN still require careful tuning of hyperparameters like learning rate and batch size. Although pretrained models can get accurate results with small datasets if the dataset is too smaller or lacks variaton it still can overfit.

In this grape disease detection system each input grape leaf image is processed through the pre trained CNN model MobileNetV2. Also model outputs a multiclass prediction allocating the image to one of the grape leaf disease categories with corresponding confidence scores. Due to its accurate results, performance and ability to lean both general and domain specific features with minimum recources CNN models are the most suitable and effective technique for grape leaf disease ditection in this project.

**Clustering (K-Means)**

Clustering in unsupervised machine learning method that works using unlabled data. Unlike supervised machine learning methods this learns without labled data. Clustering algorithms identify patterns in the data and then group those data points with similar patterns together. k-means clustering is one of most used algorithm of unspervised algorithms. K means number of clusters that defined by parameter k. In this project k can be 4 by representing potential groupings healthy, black measles, black rot and isariopsis leaf spot. In k-means algorithms works as initializing cluster centroids then using distance metrics to iteratively assigning data points to the nearest centroid. The centroids will update until convergence is achieved.

Cluestering is advantages if labled datasets are hard to obtain. Also clustering could help to uncover hidden patters within the data in the unlabled dataset like variation within disease classes or subtypes. But since clustering does not produce outputs as labled data there is a repuirement for expert knowledge to interprete which cluter corresponds to which data. Since we dont have expert knowledge and then evaluating the trained data grouping will be a difficulity and the results will not be certain therfore using clustering in grape disease detection will be less accurate than other techniques.

**Classification (Random Forest)**

Random forest is a supervised machine learning method that uses for classification tasks. It combines the predictions of multiple decision trees. Each tree decides on a class individually and then the final prediction is made using majority vote from all the trees. This group decsion making is more accurate and less prone to overfitting because it relying one multiple decision tress instead of one. The method is known for being somewhat interpretable.

A key limitation of random forest method is that it can not work with image files. Images must be converted to set of numerical features to be used in grape leaf disease detection. Traditional image proccessing methods can be used to extract thes eimages or by using the learned features from pre trained CNN model. Features can be used as input for the random forest model after the extract them and normalize them. Then the model outputs a predicted disease class for every image one by one and then provide confidence score for each possible calss. But if the features are noisy or irrelevent the performance can be dropeed. Also building many trees with many features can be computationaly expensive. Random forest usually performs less well when compare to convolutional neural networks for image tasks because CNNs learn visual patterns directly from pixels but random forests only depends on pre  extracted features. Therfore random forest can not be used to detect grape diseases as main source but it can be used as secondary classifer to analize feature importance.

```python
# import libraries

import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
```

```python
# import training and validation dataset data

train_dir = '/content/drive/MyDrive/Colab Notebooks/GrapeDataset/train'
val_dir   = '/content/drive/MyDrive/Colab Notebooks/GrapeDataset/valid'

# Dataset path and parameters

IMG_SIZE = 224
BATCH_SIZE = 32
EPOCHS = 15
NUM_CLASSES = 4
```

```python
# data augmentation and loading


train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=25,
    zoom_range=0.2,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True
)

val_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input
)

test_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input
)

```

**Image count for "Training"**

Class "Black Measles" : 3026 images

Class "Black Rot" : 2877 images

Class "Isariopsis Leaf Spot" : 2582 images

Class "Healthy" : 2075 images

**Image count for "Validation"**

Class "Black Measles" : 379 images

Class "Black Rot" : 360 images

Class "Isariopsis Leaf Spot" : 260 images

Class "Healthy" : 323 images


```python
# display a batch of augmented training images

images, labels = next(train_generator)
plt.figure(figsize=(10, 10))
for i in range(9):
    plt.subplot(3, 3, i+1)
    plt.imshow(images[i])
    plt.axis('off')
plt.show()
```

```python
# load pre-trained MobileNetV2 model

base_model = MobileNetV2(
    input_shape=(224,224,3),
    include_top=False,
    weights='imagenet'
)

base_model.trainable = False  

```

```python
# add custom classificatoin layers

x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(128, activation='relu')(x)
x = Dropout(0.5)(x)
output = Dense(NUM_CLASSES, activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=output)
```

```python
# compile the model

model.compile(
    optimizer=Adam(learning_rate=0.0001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

```python
# train the model

history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=EPOCHS
)
```

```python
# get the model summary

model.summary()
```

```python
# display accuracy and loss for train and validation

train_loss, train_acc = model.evaluate(train_generator)
val_loss, val_acc = model.evaluate(val_generator)

print(f"Train Accuracy: {train_acc:.4f}")
print(f"Train Loss: {train_loss:.4f}")
print(f"Validation Accuracy: {val_acc:.4f}")
print(f"Validation Loss: {val_loss:.4f}")
```

```python
# get predictions for the test set

test_generator.reset()
predictions = model.predict(test_generator)
y_pred_classes = np.argmax(predictions, axis=1)


y_true = test_generator.classes
class_labels = list(test_generator.class_indices.keys())
```

```python
# compute confusion matrix

cm = confusion_matrix(y_true, y_pred_classes)


plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_labels, yticklabels=class_labels)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()
```
